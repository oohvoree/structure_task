```python
class Student:
    pass

n = int(input("Кількість студентів: "))
students = []

for i in range(n):
    s = Student()
    s.name = input("ПІП: ")
    s.course = input("Курс: ")
    s.group = input("Група: ")
    s.residence = input("Місце проживання: ")
    students.append(s)

print("\nСтуденти, які проживають на квартирі:")
for s in students:
    if s.residence.lower() == "квартира":
        print(s.name, s.course, s.group)
```
