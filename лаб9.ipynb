```python
class Student:
    pass

n = int(input(" : "))
students = []

for i in range(n):
    s = Student()
    s.name = input(": ")
    s.course = input(": ")
    s.group = input(": ")
    s.residence = input(" : ")
    students.append(s)

print("\n,    :")
for s in students:
    if s.residence.lower() == "":
        print(s.name, s.course, s.group)
```
