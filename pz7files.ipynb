**datetime** ‚Äî —Ü–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∞ –≤ Python –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ –¥–∞—Ç–æ—é —Ç–∞ —á–∞—Å–æ–º.


````python
from datetime import datetime
now = datetime.now().strftime("%d.%m.%Y %H:%M")
print(f"–î–∞—Ç–∞: {now}")
````

**datetime.now()** ‚Äî –æ—Ç—Ä–∏–º—É—î –ø–æ—Ç–æ—á–Ω—É –¥–∞—Ç—É –π —á–∞—Å.

.**strftime**(...) ‚Äî —Ñ–æ—Ä–º–∞—Ç—É—î –¥–∞—Ç—É –π —á–∞—Å —É –∑—Ä—É—á–Ω–∏–π –¥–ª—è —á–∏—Ç–∞–Ω–Ω—è –≤–∏–≥–ª—è–¥:

**%d** ‚Äî –¥–µ–Ω—å

**%m** ‚Äî –º—ñ—Å—è—Ü—å

**%Y** ‚Äî —Ä—ñ–∫

**%H:%M** ‚Äî –≥–æ–¥–∏–Ω–∏ —ñ —Ö–≤–∏–ª–∏–Ω–∏

````python
from datetime import datetime

# –ö–æ—à–∏–∫
basket = []

print("üîç –î–æ–¥–∞–≤–∞–Ω–Ω—è —Ç–æ–≤–∞—Ä—ñ–≤ –¥–æ –∫–æ—à–∏–∫–∞ (–≤–≤–µ–¥—ñ—Ç—å '–∫—ñ–Ω–µ—Ü—å', —â–æ–± –∑–∞–≤–µ—Ä—à–∏—Ç–∏)")

while True:
    name = input("–ù–∞–∑–≤–∞ —Ç–æ–≤–∞—Ä—É: ")
    if name.lower() == "–∫—ñ–Ω–µ—Ü—å":
        break
    price = float(input(f"–¶—ñ–Ω–∞ —Ç–æ–≤–∞—Ä—É '{name}': "))
    countTov = int(input(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ç–æ–≤–∞—Ä—É '{name}': "))

    # –î–æ–¥–∞—î–º–æ —Ç–æ–≤–∞—Ä —É –∫–æ—à–∏–∫
    basket.append({"name": name, "price": price, "countTov": countTov})

# –î—Ä—É–∫ —á–µ–∫—É
now = datetime.now().strftime("%d.%m.%Y %H:%M")
print("\nüßæ –¢–û–í–ê–†–ù–ò–ô –ß–ï–ö")
print(f"–î–∞—Ç–∞: {now}")
print("-" * 40)

sum = 0
for item in basket:
    line_sum = item["price"] * item["countTov"]
    sum += line_sum
    print(f"{item['name']:<20} x{item['countTov']:>2}  {item['price']:>6.2f} –≥—Ä–Ω = {line_sum:>6.2f} –≥—Ä–Ω")

print("-" * 40)
print(f"{'–ó–∞–≥–∞–ª—å–Ω–∞ —Å—É–º–∞':<30} = {sum:.2f} –≥—Ä–Ω")
print("–î—è–∫—É—î–º–æ –∑–∞ –ø–æ–∫—É–ø–∫—É! üõí")

# –ó–±–µ—Ä–µ–∂–µ–Ω–Ω—è —á–µ–∫—É —É —Ñ–∞–π–ª üßæ chek.txt üõí
??????????????????????????????????????
````
